# **C. Looping**

## 1. Looping *'for'*

```python
df
```

```python
# looping index dan nilai yang ada dalam kolom "gpa", kemudian filter hanya
# nilai yang diatas 9,7

for index, nilai in df.iterrows():
    if nilai["gpa"] > 9.7:
        print(f"Index: {index}, GPA: {nilai['gpa']}")
```

```python
# menampilkan kolom yang hanya huruf terakhirnya adalah vokal

vokal = ['a','i', 'u', 'e', 'o']

# Looping kolom
# loop 1 adalah col = 'gre_score'
for col in df.columns:
    if col[-1] in vokal:
        print(f"Huruf terakhir kolom '{col}' adalah vokal")
```

## 2. Looping *'while'*

```python
# kita ingin menampilkan angka 1-50 dimana jika angkanya kelipatan 10,
# kita print angka ini kelipatan 10

i = 1

while i <= 50:
  if i % 10 == 0:
    print('Angka ini kelipatan 10')
  else:
    print(i)

  i += 1   # increment
```

## Loop Control Statement

Introduce to `break, pass, continue`

```python
i = 0

while i <= 20:
    if i == 3 or i == 7 :
        i += 1
        continue      # Lewati angka 3

    elif i == 5:
        pass          # Tidak melakukan apa pun untuk 5

    elif i == 15:
        break         # Hentikan perulangan ketika i == 15

    print(i)
    i += 1
```

| Keyword    | Fungsi                                                           |
| ---------- | ---------------------------------------------------------------- |
| `break`    | Menghentikan loop sepenuhnya                                     |
| `continue` | Melewati 1 iterasi, lanjut ke iterasi berikutnya                 |
| `pass`     | Lanjut ke kode selanjutnya. Digunakan ketika kamu butuh menulis sesuatu, tapi belum ingin mengisinya.


# **D. Lambda Function**

Kita harus menggunakan `.apply()` jika ingin menggunakan lambda di dataFrame

```python
'''
  lambda function untuk membuat kolom baru 'lolos'.
  isi dari kolom tersebut adalah "ya" atau "tidak"
  jika gre_score > 320 maka "ya", jika kurang dari itu
  maka "tidak"
'''

lamb = lambda x: 'ya' if x > 320 else 'tidak'

df['lolos'] = df['gre_score'].apply(lambda x: 'ya' if x > 320 else 'tidak')
```

```python
df.head()
```

```python
'''
  Lambda function untuk memfilter kolom 'research_exp'
  yang berawalan huruf 'y'
'''
filt = lambda x: x.startswith('y') # outputnya hanya False atau True

filter_df = df[df['research_exp'].apply(filt)]
```

```python
filter_df.head()
```

```python
# mengecek nilai unik dari kolom 'research_exp'
# .unique() = syntax untuk mencari nilai yang unik

print(filter_df['research_exp'].unique())
```

```python
print(df['research_exp'].unique())
```

```python
'''
  Lambda function untuk membuat kolom baru 'total_nilai'
  yang isinya adalah jumlah dari 'gre_score' dan 'toefl_score'

  axis = 1 digunakan untuk menjumlahkan nilai dari baris
'''

df['total_nilai'] = df.apply(lambda x: x['gre_score'] + x['toefl_score'], axis=1)
```
