1. Lakukan investigasi pada kondisi berikut :

    1a. Lakukan pengecekan pada kolom mana yang terdapat data continue di dataframe ? (20 poin)
    1b. Lakukan pengecekan pada kolom mana yang terdapat jenis kategorikal serta Uraikan juga jenis kategorikalnya termasuk kategorikal apa di masing-masing kolom dataframe ? (20 poin)


```python
#1a
data_continue = []
for col in df.columns:
  if df[col].dtype in ['float64']:
    data_continue.append(col)

print('Kolom berupa data continue:', data_continue)

#1b
data_kategorik = []
for col in df.columns:
  if df[col].dtype == 'object':
    data_kategorik.append(col)

data_kategorik.extend(['pclass', 'survived'])

print(f'Kolom kategorikal (tipe object/teks):{data_kategorik}')

jenis_kategorik = {
    'name': 'nominal',
    'sex': 'nominal',
     'ticket': 'nominal',
     'cabin': 'nominal',
     'embarked': 'nominal',
     'boat': 'nominal',
     'home.dest': 'nominal',
     'pclass': 'ordinal',
     'survived': 'nominal',
}

print(f'Jenis kategorikal: {jenis_kategorik}')



```

```python
# 2a. for loop
# Kolom yang dimulai dengan huruf konsonan
vokal = ['a', 'i', 'u', 'e', 'o']

# Looping kolom
for col in df.columns:
  huruf_konsonan = col[0]


    # Cek apakah huruf pertama adalah alfabet dan bukan huruf vokal
  if huruf_konsonan not in vokal:
    print(col)
```

```python
# 2b. while loop
# Print semua angka ganjil hingga 30 yang habis dibagi 3
i = 1
while i <= 30:
  if i % 2 == 1:
    if i % 3 == 0:
      print(f'{i} adalah angka ganjil di bawah 30 yang habis dibagi 3')
  i += 1
```

```python
# 3a. Hitung jumlah missing value (NaN) pada kolom body

# Menggunakan .apply() dengan lambda function pada kolom 'body'
# pd.isna(x) akan mengembalikan True jika nilai x adalah NaN
# .sum() akan menjumlahkan semua nilai True (True=1, False=0)
if 'body' in df.columns:
  missing_value_body = df['body'].apply(lambda x:1 if pd.isna(x) else 0).sum()
  print(f'Jumlah missing value pada kolom body sejumlah {missing_value_body}')
```

```python
# Ubah female jadi 0 dan male 1
if 'sex' in df.columns:
  df['sex'] = df['sex'].apply(lambda x: 0 if x == 'female' else x)
  df['sex'] = df['sex'].apply(lambda x: 1 if x == 'male' else x)
  print (df['sex'])
```
